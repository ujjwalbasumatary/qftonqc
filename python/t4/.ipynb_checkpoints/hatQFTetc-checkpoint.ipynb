# Quantum Fourier Transform

The Quantum Fourier Transform $\hat{QFT}$ acts as follows
\begin{equation}
\hat{QFT} \ket{k} = \frac{1}{2^{N/2}}\sum_{j = 0}^{2^N - 1} e^{2\pi i j k/2^N} \ket{j}
\end{equation}

The qiskit implementation for this is in qiskit.circuit.library.QFTGate

```python
from qiskit.circuit.library import QFTGate
from qiskit import QuantumCircuit
```

```python
num_qubits = 10
```

```python
qc = QuantumCircuit(num_qubits)
```

```python
qc.append(QFTGate(num_qubits), range(num_qubits))
```

```python
from qiskit.quantum_info import Statevector
```

```python
state = Statevector.from_instruction(qc)
```

```python
state
```

The in state above was just $\ket{0\cdots 0}$.

```python
import numpy as np
```

Numpy Discrete Fourier Transform
$$A_k = \sum_{m=0}^{n-1} a_m \exp\left\{-2\pi i{mk \over n}\right\} \qquad k = 0,\ldots,n-1$$

```python
1 / np.sqrt(2 ** num_qubits)
```

```python
in_state = np.array([1 if i == 0 else 0 for i in range(2 ** num_qubits)])
```

```python
numpy_fft = np.fft.ifft(in_state, norm='ortho')
```

```python
qiskit_fft = state.data
```

```python
np.allclose(qiskit_fft, numpy_fft)
```

```python
qc = QuantumCircuit(num_qubits)
```

```python
random_init_state = np.random.random(2 ** num_qubits)
```

```python
random_init_state /= np.sqrt(np.sum(np.abs(random_init_state)**2))
```

```python
qc.initialize(random_init_state)
```

```python
qc.append(QFTGate(num_qubits), range(num_qubits))
```

```python
out_state_qiskit = Statevector.from_instruction(qc)
```

```python
out_state_qiskit.data
```

```python
out_state_numpy = np.fft.ifft(random_init_state, norm='ortho')
```

```python
np.allclose(out_state_qiskit, out_state_numpy)
```

# Phase estimation

Let's try to use phase estimation on the $R_z$-gate, i.e. the Hamiltonian
\begin{equation}
H = J Z\, \quad J \in \mathbb{R}
\end{equation}
I will set $J = 2 \pi \tilde{J}$ since the phase estimation algorithm picks out the phase with a factor of $2\pi$ pulled out. Let's set $\tilde{J} = 0.5$.

```python
from qiskit.circuit.library import phase_estimation, RZGate
```

```python
num_eval_bits = 5 # number of qubits in the phase estimation register
num_qubits = 1 # number of qubits on which the unitary acts
```

```python
from qiskit import QuantumRegister, ClassicalRegister
```

```python
qreg = QuantumRegister(num_qubits, 'q')
evalreg = QuantumRegister(num_eval_bits, 'e')
creg = ClassicalRegister(num_eval_bits, 'c')
```

```python
qc = QuantumCircuit(evalreg, qreg, creg) # actual circuit
```

```python
qc.draw('mpl')
```

```python
tilde_J = -0.3
```

```python
qpe_circuit = phase_estimation(num_eval_bits, RZGate(-4 * np.pi * tilde_J))
```

```python
qpe_circuit.draw('mpl')
```

```python
qc.append(qpe_circuit, evalreg[:] + qreg[:])
```

```python
qc.decompose().draw('mpl')
```

```python
qc.measure(evalreg, creg)
```

```python
qc.draw('mpl')
```

```python
from qiskit.primitives import StatevectorSampler
```

```python
sampler = StatevectorSampler()
```

```python
job = sampler.run([(qc,[])], shots = 100)
```

```python
result = job.result()[0].data.c
```

```python
result.get_counts()
```

```python
def bin_to_dec(bin_string):
    bin_string = bin_string[::-1]
    ans = 0.0
    for idx, char in enumerate(bin_string):
        ans += float(char) / 2 ** (idx + 1)
    return ans
```

```python
bin_to_dec('01101')
```

Can extract eigenvalues between $[0, 1)$

# Hadamard Test

Evolve with
$$ H = 2 J Z$$

Given a unitary $U$ and a state $\ket{\psi}$ the Hadamard test estimates $\langle \psi | U |\psi \rangle \in \mathbb{C}$. Let's try to implement this with
\begin{equation}
U = R_z(\theta) = e^{-i\theta Z/2} = \begin{pmatrix} e^{-i\theta/2} & 0 \\ 0 & e^{i\theta/2} \end{pmatrix}
\end{equation}
Hence for the computational basis states $\ket{0}, \ket{1}$, we have
\begin{equation}
\begin{split}
\langle 0 | U | 0 \rangle &= e^{-i\theta/2}\\
\langle 1 | U | 1 \rangle &= e^{i\theta/2}
\end{split}
\end{equation}

```python
J = 0.3
```

```python
c_rz_gate = RZGate(J).control(1)
```

```python
qc = QuantumCircuit(2)
```

```python
qc.h(0)
qc.append(c_rz_gate, [0,1])
qc.h(0)
```

```python
qc.draw('mpl')
```

```python
creg = ClassicalRegister(1, 'c')
```

```python
qc.add_register(creg)
```

```python
qc.draw('mpl')
```

```python
qc.measure(0, creg)
```

```python
qc.draw('mpl')
```

```python
job = sampler.run([(qc, [])], shots = 100)
```

```python
job.result()[0].data.c.get_counts()
```

P(0) should be
$$P(0) = \frac{1}{2} \left( 1 + \cos \theta / 2 \right)$$

```python
0.5 * (1 + np.cos(J / 2))
```

```python
qc = QuantumCircuit(2)
```

```python
qc.h(0)
qc.sdg(0)
qc.append(c_rz_gate, [0,1])
qc.h(0)
```

```python
qc.draw('mpl')
```

```python
qc.add_register(creg)
```

```python
qc.measure(0, creg)
```

```python
job = sampler.run([(qc, [])], shots = 100)
```

```python
job.result()[0].data.c.get_counts()
```

Here, $P(0)$ should be
$$ P(0) = \frac{1}{2} \left( 1 - \sin \theta / 2\right) $$

```python
0.5*( 1- np.sin(0.3/2))
```

```python
from qiskit.circuit import Parameter
```

```python
c_rz_gate = RZGate(Parameter('theta')).control(1)
```

```python
qc_real = QuantumCircuit(2)
qc_real.h(0)
qc_real.append(c_rz_gate, [0,1])
qc_real.h(0)
```

```python
qc_real.draw('mpl')
```

```python
qc_real.add_register(creg)
```

```python
qc_real.measure(0, creg)
```

```python
sampler_real = StatevectorSampler()
```

```python
theta_vals = np.arange(0, 2 * np.pi, 0.1)
```

```python
num_shots = 10000
```

```python
job_real = sampler_real.run([(qc_real, theta_vals)], shots = num_shots)
```

```python
lst_real = []
```

```python
length = len(theta_vals)
```

```python
for i in range(length):
    result = job_real.result()[0].data.c.get_counts(i)
    lst_real.append(result.get('0', 0))
```

```python
p_real_0 = np.array(lst_real) / num_shots
```

```python
p_real_0_true = 0.5 * ( 1 + np.cos(theta_vals / 2))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(theta_vals, p_real_0, label='QC')
plt.plot(theta_vals, p_real_0_true, label='Expected')
plt.legend()
```
