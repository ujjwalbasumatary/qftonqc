# Circuits to prepare bell states

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.primitives import StatevectorSampler
from qiskit.quantum_info import Statevector
```

```python
def Phi_plus(qreg, creg):
    circuit = QuantumCircuit(qreg, creg)
    circuit.h(0)
    circuit.cx(0, 1)
    circuit.measure(qreg, creg)
    return circuit
```

```python
qreg = QuantumRegister(2, 'q')
creg = ClassicalRegister(2, 'c')
```

```python
qc = Phi_plus(qreg, creg)
```

```python
qc.draw('mpl')
```

# Statevector sampler example with two-site Ising

```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.circuit import Parameter
import numpy as np
```

## Single qubit precession

We will consider evolution under

\begin{equation}
H = Z
\end{equation}

for the state $\ket{+}$.

```python
qreg = QuantumRegister(1, 'q')
creg = ClassicalRegister(1, 'c')
```

```python
qc = QuantumCircuit(qreg, creg)
```

```python
qc.h(0)
qc.rz(Parameter('theta'), 0)
qc.h(0)
qc.measure(qreg, creg)
```

```python
qc.draw('mpl')
```

```python
sampler = StatevectorSampler()
```

```python
theta_vals = np.linspace(0, 10, 100)
```

```python
job = sampler.run([(qc, theta_vals)], shots = 100)
```

```python
results = job.result()[0]
```

```python
lst_m0 = []
lst_m1 = []
```

```python
for i in range(len(theta_vals)):
    result = results.data.c.get_counts(i)
    lst_m0.append(result.get('0', 0))
    lst_m1.append(result.get('1', 0))
```

```python
p0 = np.array(lst_m0) / 100
p1 = np.array(lst_m1) / 100
```

```python
plt.plot(theta_vals, p0)
```
