For measurements you need to put classical registers in your quantum circuit

```python
from qiskit.circuit import ClassicalRegister, QuantumRegister
```

```python
qreg = QuantumRegister(2, 'q')
creg = ClassicalRegister(2, 'c')
```

```python
qc_bell = QuantumCircuit(qreg, creg)
```

```python
qc_bell.draw('mpl')
```

```python
qc_bell.h(0)
qc_bell.cx(0, 1)
qc_bell.measure(qreg, creg)
```

```python
qc_bell.draw('mpl')
```

```python
from qiskit.primitives import StatevectorSampler
```

```python
sampler = StatevectorSampler()
```

```python
num_shots = 10000
```

```python
job = sampler.run([(qc_bell, [])], shots = num_shots)
```

```python
results = job.result()[0]
```

```python
results
```

```python
results.data.c.get_counts()
```

# Single-qubit precession

Implement time evolution on just one qubit with

\begin{equation}
H = Z
\end{equation}

```python
qreg = QuantumRegister(1, 'q')
creg = ClassicalRegister(1, 'c')
```

```python
qc_precession = QuantumCircuit(qreg, creg)
```

```python
qc_precession.draw('mpl')
```

```python
from qiskit.circuit import Parameter
```

```python
qc_precession.h(0)
qc_precession.rz(Parameter('theta'), 0)
qc_precession.h(0)
```

```python
qc_precession.draw('mpl')
```

```python
qc_precession.measure(qreg, creg)
```

```python
qc_precession.draw('mpl')
```

```python
theta_vals = np.linspace(0, 10, 100)
```

```python
sampler = StatevectorSampler()
```

```python
num_shots = 5
```

```python
job = sampler.run([(qc_precession, theta_vals)], shots = num_shots)
```

```python
lst_m0 = []
lst_m1 = []
```

```python
results = job.result()[0]
```

```python
for i in range(len(theta_vals)):
    result = results.data.c.get_counts(i)
    lst_m0.append(result.get('0', 0))
    lst_m1.append(result.get('1', 0))
```

```python
p0 = np.array(lst_m0) / num_shots
p1 = np.array(lst_m1) / num_shots
```

```python
import matplotlib.pyplot as plt
```

```python
exact = np.cos(theta_vals / 2) ** 2
```

```python
plt.plot(theta_vals, p0)
plt.plot(theta_vals, exact)
```

# expectation values

```python
qc_expect = QuantumCircuit(1)
```

```python
qc_expect.draw('mpl')
```

```python
qc_expect.h(0)
qc_expect.rz(Parameter('theta'), 0)
qc_expect.h(0)
```

```python
qc_expect.draw('mpl')
```

```python
from qiskit.primitives import StatevectorEstimator
```

```python
estimator = StatevectorEstimator()
```

```python
from qiskit.quantum_info import SparsePauliOp
```

```python
op = SparsePauliOp('Z')
```

```python
op
```

```python
job = estimator.run([(qc_expect, op, theta_vals)])
```

```python
results = job.result()[0]
```

```python
exp_val = results.data.evs
```

```python
exact = np.cos(theta_vals)
```

```python
plt.plot(theta_vals, exp_val)
plt.plot(theta_vals, exact)
```

# two-site Ising

The hamiltonian for the two site Ising model is
\begin{equation}
H = Z \otimes Z
\end{equation}

```python
qreg = QuantumRegister(2, 'q')
creg = ClassicalRegister(2, 'c')
```

```python
qc_ising = QuantumCircuit(qreg, creg)
```

```python
qc_ising.h([0, 1])
```

```python
qc_ising.draw('mpl')
```

```python
qc_ising.cx(0, 1)
```

```python
qc_ising.rz(2 * Parameter('J'), 1)
```

```python
qc_ising.cx(0,1)
```

```python
qc_ising.h([0,1])
```

```python
qc_ising.draw('mpl')
```

```python
qc_ising.measure(qreg, creg)
```

```python
qc_ising.draw('mpl')
```

```python
sampler = StatevectorSampler()
```

```python
theta_vals = np.linspace(0, 10, 100)
```

```python
num_shots = 100
```

```python
job = sampler.run([(qc_ising, theta_vals)], shots = num_shots)
```

```python
results = job.result()[0]
```

```python
results.data.c.get_counts(50)
```

```python
lst_m00 = []
lst_m11 = []
for i in range(len(theta_vals)):
    result = results.data.c.get_counts(i)
    lst_m00.append(result.get('00', 0))
    lst_m11.append(result.get('11', 0))
```

```python
p00 = np.array(lst_m00) / num_shots
p11 = np.array(lst_m11) / num_shots
```

```python
exact = np.sin(theta_vals) ** 2
```

```python
plt.plot(theta_vals, p11)
plt.plot(theta_vals, exact)
```
